```python
Hate speech detection:
the goal of the project was to classify given text into 3 categories-
"hate speech","offensive" and "neither of them". I used a twitter dataset
which contained nearly 25000 labeled tweets to train and test the model. I implemented
transfer learning on BERT to build the model. The model was
of pre-trained BERT architecture with additional layers appended on it, along
with softmax activation on output layer for multiclass classification.
Moreover, I used Adam optimizer for speeding the training process. The model
attained an accuracy of 93% on the test dataset.
```
