```python
def param_init(Y):
  m=0.1
  c=Y.mean()
  return m,c
```

```python
def generate_prediction(m,c,X):
  prediction=[]
  for i in X:
    prediction.append(m*i+c)
  return prediction
```

```python
from sklearn.metrics import mean_squared_error
def cost_f(prediction,Y):
  MSE=mean_squared_error(prediction,Y)
  return MSE
```

```python
def gradients(prediction,Y,X):
  Gm=np.sum((prediction-Y)*X)*2/len(Y)
  Gc=np.sum((prediction-Y))*2/len(Y)
  return Gm,Gc
  
```

```python
def update_param(m_old,c_old,Gm,Gc,alpha):
  m_new=m_old-alpha*Gm
  c_new=c_old-alpha*Gc
  return m_new,c_new
```

```python
def result(m,c,X,Y,cost,prediction,i):
  if i < max_iter-1:
    print("Gradient has been converged on {} iteration".format(i))
  else:
    print(max_iter)
  plt.scatter(X,Y,color="red")
  plt.plot(X,prediction,color="green")
  print("m = {} and c = {}".format(m,c))
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()

sale_price=scaler.fit_transform(sample_data["Sale_Price"].values.reshape(-1,1))
flat_area=scaler.fit_transform(sample_data["flat_area"].values.reshape(-1,1))

alpha=0.01
max_iter=1000
cost_old=0

m,c= param_init(sale_price)
for i in range(0,max_iter):
    prediction= generate_prediction(m,c,flat_area)
    cost_new=cost_f(prediction,flat_area)
    if (cost_old-cost_new)<10**(-7):
      break
    Gm,Gc=gradients(prediction,sale_price,flat_area)
    m,c=update_param(m,c,Gm,Gc,alpha)
    if i%20==0:
      print(" iteration "+str(i)+" m = "+str(m)+" c = "+str(c)+" cost = "+str(cost_new))
    cost_old=cost_new

result(m,c,flat_area,sale_price,cost_new,prediction,i)
```

## **Calculting VIF**

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif_data=X
VIF=pd.Series([variance_inflation_factor(vif_data.values,i) for i in range(vif_data.shape[1])],index=vif_data.columns)
VIF
```

```python
VIF[VIF==VIF.max()].index[0]
```

```python
def MC_remover(data):
  vif=pd.Series([variance_inflation_factor(data.values,i) for i in range(data.shape[1])],index=data.columns)
  if vif.max()>5:
    print(vif[vif==vif.max()].index[0], "has been droped")
    data=data.drop(columns=vif[vif==vif.max()].index[0])
    return data
  else:
    print("no multicollinearity present")
    return data
```

```python
for i in range(7):
  vif_data=MC_remover(vif_data)
vif_data.head()
```

```python
VIF=pd.Series([variance_inflation_factor(vif_data.values,i) for i in range(vif_data.shape[1])],index=vif_data.columns)
VIF
```

```python
x=vif_data
y=data["Sale_Price"]
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.3,random_state=101)
```

# **Linear Regression**

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression(normalize=True)
lr.fit(X_train,Y_train)
```

```python
prediction=lr.predict(X_test)
R_square=lr.score(X_test,Y_test)
R_square
```
